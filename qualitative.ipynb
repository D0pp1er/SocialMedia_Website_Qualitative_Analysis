```python
import pandas as pd
import numpy as np
data_path = 'qualitative - Asib.csv'

data = pd.read_csv(data_path)
data.fillna(0, inplace=True)
```

```python
data.columns
```

```python
def get_codes_count(name):

    codes = {}
    for i in data[name].values:
        if i == 0:
            continue
        code = i.split(',')
        
        for j in code:
            j = j.strip()
            if j not in codes:
                codes[j] = 1
            else:
                codes[j] += 1
    
    # for k in codes.keys():
    #     print(f'{k},{codes[k]}')
    return codes
    
```

```python
all_codes = ['Code.1','Code.2','Code.3','Code.4','Code.5','Code.6','Code.7','Code.8']

# data[all_codes]
code = []
mx_len = 0
for i in all_codes:
    code.append( get_codes_count(i) )
    mx_len = max(mx_len, len(code[-1].keys()))

print('\n'.join([f'{code[0][k]}' for k in code[0].keys()]))
print('\n'.join([f'{k}' for k in code[0].keys()]))
```
